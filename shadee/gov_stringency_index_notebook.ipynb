```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import json
import requests
```

```python
url = 'https://raw.githubusercontent.com/OxCGRT/USA-covid-policy/master/data/OxCGRT_US_subnational_05Aug2020.csv'
res = requests.get(url, allow_redirects=True)
with open('covid_stringency.csv','wb') as file:
    file.write(res.content)
covid_index_df = pd.read_csv('covid_stringency.csv', low_memory=False)
```

```python
covid_index_df
```

```python
#Groupby State
state_stringency_index_groupby = covid_index_df.groupby(['RegionName'])

#Average Stringency
state_stringency_index_for_display = state_stringency_index_groupby["StringencyIndexForDisplay"].mean()

#Legacy Index
state_stringency_legacy_index_display = state_stringency_index_groupby["StringencyLegacyIndexForDisplay"].mean()

# Government Response
state_government_response = state_stringency_index_groupby["GovernmentResponseIndexForDisplay"].mean()

#Containment Health Index
containment_health_index = state_stringency_index_groupby["ContainmentHealthIndexForDisplay"].mean()

#Economic Support Index
economic_support_index = state_stringency_index_groupby["EconomicSupportIndexForDisplay"].mean()


state_by_state_df = pd.DataFrame({
    "Average Stringency Index": state_stringency_index_for_display,
    "Average Stringency Legacy Index": state_stringency_legacy_index_display,
    "Average Government Response": state_government_response,
    "Average Containment Health Risk": containment_health_index,
    "Average Economic Support Index": economic_support_index
    
}).reset_index()

state_by_state_df.reset_index()
```

```python
#Include only cities within the alcohol data

cleaned_state_by_state_df = state_by_state_df.loc[(state_by_state_df["RegionName"] == 'Alaska') | (state_by_state_df["RegionName"] == 'Arkansas') |
                                                   (state_by_state_df["RegionName"] == 'Colorado') | (state_by_state_df["RegionName"] == 'Connecticut')
                                                  | (state_by_state_df["RegionName"] == 'Delaware') | (state_by_state_df["RegionName"] == 'Florida') |
                                                  (state_by_state_df["RegionName"] == 'Illinois') | (state_by_state_df["RegionName"] == 'Kansas') | 
                                                  (state_by_state_df["RegionName"] == 'Kentucky') | (state_by_state_df["RegionName"] == 'Louisiana') |
                                                  (state_by_state_df["RegionName"] == 'Massachusetts') | (state_by_state_df["RegionName"] == 'Missouri') |
                                                  (state_by_state_df["RegionName"] == 'North Dakota') | (state_by_state_df["RegionName"] == 'Oregon') | 
                                                  (state_by_state_df["RegionName"] == 'Tennessee') | (state_by_state_df["RegionName"] == 'Texas') | 
                                                  (state_by_state_df["RegionName"] == 'Utah') | (state_by_state_df["RegionName"] == 'Virginia') |
                                                  (state_by_state_df["RegionName"] == 'Wisconsin')]


cleaned_state_by_state_df.to_csv(r'cleaned_covid_stringency_index.csv', index = False)

cleaned_state_by_state_df
```
